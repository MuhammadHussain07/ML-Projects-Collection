```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('hiring.csv')
df
```

```python
import math
median_test_score=math.floor(df['test_score(out of 10)'].mean())
median_test_score
```

```python
df['test_score(out of 10)']=df['test_score(out of 10)'].fillna(median_test_score)
df
```

```python
# prompt: Using dataframe df: experience column fill NaN with median but mapping first

# Create a mapping dictionary for experience levels
experience_mapping = {
    'five':5, 'two':2, 'seven':7, 'three':3, 'ten':10, 'eleven':11,
    0: 0,  # Example mapping, replace with your actual mappings
    # Add more experience levels and their corresponding numerical values
}

# Map the experience column using the mapping dictionary
df['experience'] = df['experience'].map(experience_mapping)

# Calculate the median of the experience column
median_experience = df['experience'].median()
median_experience
# Fill NaN values in the experience column with the calculated median
df['experience']=df['experience'].fillna(median_experience)
df
```

```python
model = LinearRegression()
model.fit(df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']], df['salary($)'])
```

```python
model.predict([[2,9,6]])
```
